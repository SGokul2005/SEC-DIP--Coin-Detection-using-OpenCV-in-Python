```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
//GOKUL S
212223040051
# --- Load Image From Local Path ---
image_path = r"C:\Users\admin\Pictures\Screenshots\Screenshot 2025-11-14 083605.png"

image = cv2.imread(image_path)

if image is None:
    print("âŒ Error: Image not found. Check the file path.")
    raise SystemExit
else:
    print("âœ” Image loaded successfully")

# --- Step 1: Preprocessing ---
def preprocess_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    return blurred

# --- Step 2: Edge + Contour Detection ---
def detect_coin_edges(preprocessed, original):
    kernel = np.ones((3, 3), np.uint8)
    morph = cv2.morphologyEx(preprocessed, cv2.MORPH_CLOSE, kernel)

    edges = cv2.Canny(morph, 30, 100)

    contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    result = original.copy()
    cv2.drawContours(result, contours, -1, (0, 255, 0), 2)
    return result, edges

# --- Step 3: Display Output ---
def present_results(original_image, processed_image, edges):
    original_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    processed_rgb = cv2.cvtColor(processed_image, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(15, 6))

    plt.subplot(1, 3, 1)
    plt.title("Original Image")
    plt.imshow(original_rgb)
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("Detected Edges (Canny)")
    plt.imshow(edges, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("Coin Contours")
    plt.imshow(processed_rgb)
    plt.axis('off')

    plt.show()

# --- Run Steps ---
preprocessed = preprocess_image(image)
coin_detected_image, edges = detect_coin_edges(preprocessed, image)
present_results(image, coin_detected_image, edges)
```
